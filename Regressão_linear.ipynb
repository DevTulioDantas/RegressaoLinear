```python
x = base_casa[['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number_of_Rooms',
       'Avg._Area_Number_of_Bedrooms', 'Area_Population']]

y = base_casa[['Price']]
```

Separamos as bases.

O Objetivo é chegar ao preço.

em x, váriaveis explicativas

deixamos todas variaveis que impactam no preço no eixo X

e target em Y

e deixamos o objetivo final na base Y, o preço

Agora o machine learning

treinando o modelo pra aprender com os dados
vamos usar a biblioteca sklearn

vamos dividar a base de treino para o algoritmo aprender

e depois testar em uma base de teste


```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = 0.7,test_size = 0.3, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
y_pred = lm.predict(x_test)
```

Quanto maior o R2 SCORE maior a possibilidade de eu ter um acerto



```python
from sklearn.metrics import r2_score

r = r2_score(y_test, y_pred)
```

```python
print('r_quadrado é:', r)
```

```python
import matplotlib.pyplot as plt
```

```python
c = [i for i in range(1, 1501, 1)]

fig = plt.figure(figsize=(12,8))
plt.plot(c, y_test, color="blue")
plt.plot(c, y_pred, color="red")
plt.xlabel("index")
plt.ylabel("Preço do Imóvel")
```
